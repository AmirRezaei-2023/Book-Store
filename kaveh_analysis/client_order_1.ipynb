# Client’s Order 1

“Top Five Romance Authors”

```python
import pandas as pd
import re
```

```python
df_books = pd.read_csv("data_first_run/iraniketab.csv")

df_books.head()
```

```python
# Filter romance books and drop unnecessary columns

romance_books = df_books[df_books["category"].apply(lambda x: "عاشقانه" in x)].drop_duplicates().copy()
romance_books = romance_books[["writer", "grade", "category"]]

romance_books
```

```python
# correct writer names' format from "['نام نویسنده']" to 'نام نویسنده'


def correct_writer_name(name_string):
    match = re.search(r"\['(.*?)'\]", name_string)
    return match.group(1) if match else None


romance_books["writer"] = romance_books["writer"].apply(correct_writer_name)

romance_books["writer"].value_counts()
```

```python
# convert category column values from string to list of strings

def category_string_to_list(tag_string):
    return list([x.strip() for x in re.findall(r"\'(.*?)\'", tag_string)])

romance_books["category"] = romance_books["category"].apply(category_string_to_list)

romance_books.head()
```

```python
# Calculate F1: Number of romance books by each author

f1 = romance_books.groupby("writer").size().reset_index(name="number_of_books")

f1
```

```python
# Calculate F2: Average rating of the romance books by each author

f2 = romance_books.groupby("writer")["grade"].mean().reset_index(name="avg_rating")
f2["avg_rating"] = f2["avg_rating"].round(3)

# just checking top writers by avg_rating
f2.nlargest(10, "avg_rating")
```

```python
# Merge F1 and F2

factors = f1.merge(f2, on="writer", how="inner")

factors
```

```python
# Normalize F1 and F2 to range (1, 2)

factors["number_of_books_normalized"] = 1 \
    + (factors["number_of_books"] - factors["number_of_books"].min()) \
    / (factors["number_of_books"].max() - factors["number_of_books"].min())

factors["avg_rating_normalized"] = 1 \
    + (factors["avg_rating"] - factors["avg_rating"].min()) \
    / (factors["avg_rating"].max() - factors["avg_rating"].min())

factors
```

```python
# Calculate F3: Number of award tags for author’s romance books

def count_awards(tags):
    count = 0
    for tag in tags:
        if "جایزه" in tag:
            count += 1
    return count


romance_books["award_tags_count"] = romance_books["category"].apply(count_awards)
f3 = (
    romance_books.groupby("writer")["award_tags_count"]
    .sum()
    .reset_index(name="award_tags_count")
)

f3
```

```python
# Calculate F4: Number of list tags for author’s romance books


def count_listings(tags):
    count = 0
    for tag in tags:
        if ("ترین" in tag) or ("گزیده" in tag):
            count += 1
    return count


romance_books["list_tags_count"] = romance_books["category"].apply(count_listings)
f4 = (
    romance_books.groupby("writer")["list_tags_count"]
    .sum()
    .reset_index(name="list_tags_count")
)

f4
```

```python
# Merge F3 and F4 with the previous factors dataframe

factors = factors.merge(f3, on="writer", how="inner").merge(
    f4, on="writer", how="inner")

factors
```

```python
# Normalize F3 and F4 to range (1, 2)

factors["award_tags_count_normalized"] = 1 \
    + (factors["award_tags_count"] - factors["award_tags_count"].min()) \
    / (factors["award_tags_count"].max() - factors["award_tags_count"].min())

factors["list_tags_count_normalized"] = 1 \
    + (factors["list_tags_count"] - factors["list_tags_count"].min()) \
    / (factors["list_tags_count"].max() - factors["list_tags_count"].min())

factors
```

```python
# Fill NaN values with 1.0

factors.fillna(1.0, inplace=True)

factors
```

```python
# Compute the final metric for each author
# Approach 1:
#   Very strong Rating effect

factors["final_metric_1"] = round(
    factors["number_of_books_normalized"]
    * factors["avg_rating_normalized"] ** 3
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"]
    , 3
)

factors[["writer", "final_metric_1"]].sort_values(
    by="final_metric_1", ascending=False
).head(5)
```

```python
# Compute the final metric for each author
# Approach 2:
#   Strong Rating effect

factors["final_metric_2"] = round(
    factors["number_of_books_normalized"]
    * factors["avg_rating_normalized"] ** 2
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"]
    , 3
)

factors[["writer", "final_metric_2"]].sort_values(
    by="final_metric_2", ascending=False
).head(5)
```

```python
# Compute the final metric for each author
# Approach 3: 
#   Very strong Rating effect
#   Strong number of books effect

factors["final_metric_3"] = round(
    factors["number_of_books_normalized"] ** 2
    * factors["avg_rating_normalized"] ** 3
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"]
    , 3
)

factors[["writer", "final_metric_3"]].sort_values(
    by="final_metric_3", ascending=False
).head(5)
```

```python
# Rank the authors based on the computed metric in descending order
# Choosen Approach: Approach 3

top_writers = factors[['writer', 'final_metric_3']].sort_values(by='final_metric_3', ascending=False).head(5)

top_writers
```
