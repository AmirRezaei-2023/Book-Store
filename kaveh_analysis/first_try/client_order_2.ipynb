# Client’s Order 2

“Books With Best Quality & Price”

```python
import pandas as pd
import re
```

```python
df_books = pd.read_csv("data_first_run/iraniketab.csv")

df_books.head()
```

```python
# Drop unnecessary columns
# Create cheap & quality books dataframe: cheq

cheq = df_books[
    [
        "title_persian",
        "title_english",
        "writer",
        "grade",
        "cover",
        "page_count",
        "release_year_sh",
        "category",
        "price",
    ]
].copy()

cheq
```

```python
# Handle missing values

# Assume the null values are soft cover (Hard = 1, Soft = 0)
cheq["cover"] = cheq["cover"].fillna("نرم")

# Assume the null values should be the mean of the page counts
cheq["page_count"] = cheq["page_count"].fillna(cheq["page_count"].mean())

# Assume the null values should be the median of the release years
cheq["release_year_sh"] = cheq["release_year_sh"].fillna(
    cheq["release_year_sh"].median()
)

cheq.isnull().sum()
```

```python
# Correct values of columns 'page_count', 'release_year_sh' & 'price' to be integers
cheq["page_count"] = cheq["page_count"].apply(lambda x: int(x))
cheq["release_year_sh"] = cheq["release_year_sh"].apply(lambda x: int(x))
cheq["price"] = cheq["price"].apply(lambda x: int(x))

cheq.head()
```

```python
# convert category column values from string to list of strings


def category_string_to_list(tag_string):
    return list([x.strip() for x in re.findall(r"\'(.*?)\'", tag_string)])


cheq["category"] = cheq["category"].apply(category_string_to_list)

cheq.head()
```

```python
# F1: Normalize book's rating

cheq["f1_rating"] = (cheq["grade"] - cheq["grade"].min()) / (
    cheq["grade"].max() - cheq["grade"].min()
)

cheq.head()
```

```python
# F2: Assign cover type
# Hard = 1 / covers including text "سخت" or "زرکوب" or "چرم"
# Soft = 0 / covers including text "نرم" or "سلفونی" or "شومیز"


def cover_type(cover):
    if "سخت" in cover or "زرکوب" in cover or "چرم" in cover:
        return 1
    else:
        return 0


cheq["f2_cover"] = cheq["cover"].apply(cover_type)

cheq["f2_cover"].value_counts()
```

```python
# F3: Assign page count type
# Short = 0, Medium = 0.5, Long = 1

page_bins = [
    0,
    df_books["page_count"].quantile(0.25),
    df_books["page_count"].quantile(0.75),
    df_books["page_count"].max() + 1,
]
page_labels = [0, 0.5, 1]
cheq["f3_pages"] = pd.cut(
    cheq["page_count"], bins=page_bins, labels=page_labels, include_lowest=True
)

cheq["f3_pages"] = cheq["f3_pages"].astype(float)
cheq["f3_pages"].value_counts()
```

```python
# F4: Assign publish year type
# Classic = 0.3, Old = 0.6, Modern = 1

year_bins = [
    0,
    df_books["release_year_sh"].quantile(0.33),
    df_books["release_year_sh"].quantile(0.66),
    df_books["release_year_sh"].max() + 1,
]
year_labels = [0.3, 0.6, 1]
cheq["f4_year"] = pd.cut(
    cheq["release_year_sh"], bins=year_bins, labels=year_labels, include_lowest=True
)

cheq["f4_year"] = cheq["f4_year"].astype(float)

cheq["f4_year"].value_counts()
```

```python
# F5: Number of award tags for each book
# Normalize the number of awards ( award number / maximum award number)


def count_awards(tags):
    count = 0
    for tag in tags:
        if "جایزه" in tag:
            count += 1
    return count


cheq["f5_award"] = cheq["category"].apply(count_awards)
cheq["f5_award"] = round(cheq["f5_award"] / cheq["f5_award"].max(), 2)

cheq["f5_award"].value_counts()
```

```python
# F6: Number of Listings tags for each book
# Normalize the number of listings ( listings number / maximum award listings)


def count_listings(tags):
    count = 0
    for tag in tags:
        if ("ترین" in tag) or ("گزیده" in tag):
            count += 1
    return count


cheq["f6_list"] = cheq["category"].apply(count_listings)
cheq["f6_list"] = round(cheq["f6_list"] / cheq["f6_list"].max(), 1)

cheq["f6_list"].value_counts()
```

```python
# Weights for each feature
# Approach 1: default weights
w1 = 40
w2 = 5
w3 = 5
w4 = 5
w5 = 23
w6 = 22
```

```python
# Compute the final metric for each book

cheq["metric"] = (
    w1 * cheq["f1_rating"]
    + w2 * cheq["f2_cover"]
    + w3 * cheq["f3_pages"]
    + w4 * cheq["f4_year"]
    + w5 * cheq["f5_award"]
    + w6 * cheq["f6_list"]
)

cheq.head()
```

```python
# 4. Filter books based on the criteria:
# # top quality (based on the final metric) and bottom 20% prices

threshold_price = cheq["price"].quantile(0.2)
threshold_metric = cheq["metric"].quantile(0.75)

recommended_books = cheq[
    (cheq["price"] <= threshold_price) & (cheq["metric"] >= threshold_metric)
]

recommended_books
```

```python
cleaned_recommended_books = recommended_books[
    ["title_persian", "title_english", "writer", "price", "grade", "metric"]
].sort_values(by="metric", ascending=False)

cleaned_recommended_books
```
