# Author’s Order / 3

“Top Five History Book Publishers”

```python
import pandas as pd
import re
```

```python
df_books = pd.read_csv("data_first_run/iraniketab.csv")

df_books.head()
```

```python
df_publishers = pd.read_csv("data_first_run/publisher2.csv")
df_publishers.head()
```

```python
# Filter history books and drop unnecessary columns

history_books = (
    df_books[df_books["category"].apply(lambda x: "تاریخ" in x)]
    .drop_duplicates()
    .copy()
)
history_books = history_books[["publisher_id", "grade", "category"]]

history_books
```

```python
# Correct Publishers' ID to be integers

history_books["publisher_id"] = history_books["publisher_id"].apply(
    lambda x: int(re.findall(r"\d+", x)[0])
)

history_books
```

```python
# convert category column values from string to list of strings


def category_string_to_list(tag_string):
    return list([x.strip() for x in re.findall(r"\'(.*?)\'", tag_string)])


history_books["category"] = history_books["category"].apply(category_string_to_list)

history_books.head()
```

```python
# F1: Count of historical books by publisher

f1 = history_books.groupby("publisher_id").size().reset_index(name="count_books")

f1
```

```python
# F2: Average rating of historical books by publisher

f2 = (
    history_books.groupby("publisher_id")["grade"].mean().reset_index(name="avg_rating")
)

# just checking top writers by avg_rating
f2.nlargest(10, "avg_rating")
```

```python
# Merge F1 and F2

factors = f1.merge(f2, on="publisher_id")

factors
```

```python
# Normalize F1 and F2 to range (1, 2)

factors["count_books_normalized"] = 1 + (
    factors["count_books"] - factors["count_books"].min()
) / (factors["count_books"].max() - factors["count_books"].min())

factors["avg_rating_normalized"] = 1 + (
    factors["avg_rating"] - factors["avg_rating"].min()
) / (factors["avg_rating"].max() - factors["avg_rating"].min())

factors
```

```python
# Calculate F3: Number of award tags for author’s romance books


def count_awards(tags):
    count = 0
    for tag in tags:
        if "جایزه" in tag:
            count += 1
    return count


history_books["award_tags_count"] = history_books["category"].apply(count_awards)
f3 = (
    history_books.groupby("publisher_id")["award_tags_count"]
    .sum()
    .reset_index(name="award_tags_count")
)

f3.award_tags_count.value_counts()
```

```python
# Calculate F4: Number of list tags for author’s romance books


def count_listings(tags):
    count = 0
    for tag in tags:
        if ("ترین" in tag) or ("گزیده" in tag):
            count += 1
    return count


history_books["list_tags_count"] = history_books["category"].apply(count_listings)
f4 = (
    history_books.groupby("publisher_id")["list_tags_count"]
    .sum()
    .reset_index(name="list_tags_count")
)

f4["list_tags_count"].value_counts()
```

```python
# Merge F3 and F4 with the previous factors dataframe

factors = factors.merge(f3, on="publisher_id", how="inner").merge(
    f4, on="publisher_id", how="inner"
)

factors
```

```python
# Normalize F3 and F4 to range (1, 2)

factors["award_tags_count_normalized"] = 1 + (
    factors["award_tags_count"] - factors["award_tags_count"].min()
) / (factors["award_tags_count"].max() - factors["award_tags_count"].min())

factors["list_tags_count_normalized"] = 1 + (
    factors["list_tags_count"] - factors["list_tags_count"].min()
) / (factors["list_tags_count"].max() - factors["list_tags_count"].min())

factors
```

```python
# Fill NaN values with 1.0

factors.fillna(1.0, inplace=True)

factors
```

```python
# Compute the final metric for each author
# Approach 1:
#   Very strong Rating effect

factors["final_metric_1"] = round(
    factors["count_books_normalized"]
    * factors["avg_rating_normalized"] ** 3
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"],
    3,
)

factors[["publisher_id", "final_metric_1"]].sort_values(
    by="final_metric_1", ascending=False
).head(5)
```

```python
# Compute the final metric for each author
# Approach 2:
#   Strong Rating effect

factors["final_metric_2"] = round(
    factors["count_books_normalized"]
    * factors["avg_rating_normalized"] ** 2
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"],
    3,
)

factors[["publisher_id", "final_metric_2"]].sort_values(
    by="final_metric_2", ascending=False
).head(5)
```

```python
# Compute the final metric for each author
# Approach 3:
#   Very strong Rating effect
#   Strong number of books effect

factors["final_metric_3"] = round(
    factors["count_books_normalized"] ** 2
    * factors["avg_rating_normalized"] ** 3
    * factors["award_tags_count_normalized"]
    * factors["list_tags_count_normalized"],
    3,
)

factors[["publisher_id", "final_metric_3"]].sort_values(
    by="final_metric_3", ascending=False
).head(5)
```

```python
factors
```

```python
# Rank the publishers based on the computed metric in descending order
# Choosen Approach: Approach 3

top_publishers = (
    factors[["publisher_id", "final_metric_3"]]
    .sort_values(by="final_metric_3", ascending=False)
    .head(5)
)

top_publishers_names = pd.merge(
    top_publishers, df_publishers, left_on="publisher_id", right_on="id", how="left"
).drop(columns=["id"])

top_publishers_names
```
